```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, TimeDistributed

# Step 1: Load the data
train_data = pd.read_csv("/kaggle/input/ai-for-social-good-aries-iitd-x-kaizen-24/Public/Task 2/train.csv")
test_data = pd.read_csv("/kaggle/input/ai-for-social-good-aries-iitd-x-kaizen-24/Public/Task 2/test.csv")
dev_data = pd.read_csv("/kaggle/input/ai-for-social-good-aries-iitd-x-kaizen-24/Public/Task 2/dev.csv")

# Assuming 'tokens', 'span_start_index', and 'span_end_index' are the relevant columns in your training data
tokens = train_data['tokens']
start_indices = train_data['span_start_index']
end_indices = train_data['span_end_index']

# Step 2: Tokenization and Padding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(tokens)
X = tokenizer.texts_to_sequences(tokens)
X = pad_sequences(X, padding='post')  # Pad sequences to ensure uniform length

# Step 3: Define Model Inputs and Outputs
y_start = np.array(start_indices)
y_end = np.array(end_indices)

# Step 4: Split Data into Training and Validation Sets
X_train, X_val, y_start_train, y_start_val, y_end_train, y_end_val = train_test_split(
    X, y_start, y_end, test_size=0.2, random_state=42)

# Step 5: Define Model Architecture
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100),  # Embedding layer
    Bidirectional(LSTM(units=64, return_sequences=True)),  # Bidirectional LSTM layer
    TimeDistributed(Dense(1, activation='sigmoid')),  # TimeDistributed Dense layer
])

# Step 6: Compile the Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()
```

```python
# Step 7: Train the Model
history = model.fit(X_train, y_start_train, epochs=10, batch_size=32, validation_data=(X_val, y_start_val))
```

```python
print("y_start_train shape:", y_start_train.shape)
print("y_start_val shape:", y_start_val.shape)
print("X_val shape:", X_val.shape)
print("X_train shape:", X_train.shape)
```

```python
# Reshape the target data
y_start_train = np.expand_dims(y_start_train, axis=-1)
y_start_val = np.expand_dims(y_start_val, axis=-1)

print("Reshaped y_start_train shape:", y_start_train.shape)
print("Reshaped y_start_val shape:", y_start_val.shape)
```

```python
# Train the model
history = model.fit(X_train, y_start_train, epochs=10, batch_size=32, validation_data=(X_val, y_start_val))
```
